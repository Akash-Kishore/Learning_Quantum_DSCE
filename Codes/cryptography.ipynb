```python
from cryptography.hazmat.primitives.ciphers import Cipher, algorithms, modes
from cryptography.hazmat.primitives import padding
from cryptography.hazmat.backends import default_backend
import os

# ================= ENCRYPT AUDIO FILE =================
def encrypt_file(input_path, key, output_path):
    with open(input_path, 'rb') as f:
        data = f.read()

    # Generate random IV (16 bytes for AES)
    iv = os.urandom(16)

    # PKCS7 Padding
    padder = padding.PKCS7(128).padder()
    padded_data = padder.update(data) + padder.finalize()

    # AES-CBC and quantum Encryption
    cipher = Cipher(algorithms.AES(key), modes.CBC(iv), backend=default_backend())
    encryptor = cipher.encryptor()
    cipher_text = encryptor.update(padded_data) + encryptor.finalize()

    # Save IV + Ciphertext
    with open(output_path, 'wb') as f:
        f.write(iv + cipher_text)

    print("✅ Audio encrypted successfully!")


# ================= DECRYPT AUDIO FILE =================
def decrypt_file(encrypted_path, key, output_path):
    with open(encrypted_path, 'rb') as f:
        encrypted_data = f.read()

    # Extract IV and Ciphertext
    iv = encrypted_data[:16]
    cipher_text = encrypted_data[16:]

    # AES-CBC Decryption
    cipher = Cipher(algorithms.AES(key), modes.CBC(iv), backend=default_backend())
    decryptor = cipher.decryptor()
    padded_data = decryptor.update(cipher_text) + decryptor.finalize()

    # Remove Padding
    unpadder = padding.PKCS7(128).unpadder()
    data = unpadder.update(padded_data) + unpadder.finalize()

    with open(output_path, 'wb') as f:
        f.write(data)

    print("✅ Audio decrypted successfully!")


# ================= MAIN =================
if __name__ == "__main__":

    # 🔐 SAME KEY FOR BOTH (16 bytes = AES-128)
    key = b'_Baaakaa_heehee_'

    # 🎵 Original Audio File
    input_file = r"C:\Learning_Qiskit\Files\Original\Naruto Audio.mp3"

    # 📦 Encrypted Output
    encrypted_file = r"C:\Learning_Qiskit\Files\Encrypt\Naruto Audio_Encrypt.bin"

    # 🔊 Decrypted Output
    decrypted_file = r"C:\Learning_Qiskit\Files\Decrypt\Naruto Audio_Decrypt.mp3"

    # Encrypt
    encrypt_file(input_file, key, encrypted_file)

    # Decrypt
    decrypt_file(encrypted_file, key, decrypted_file)
```

```python
from cryptography.hazmat.primitives.ciphers import Cipher, algorithms, modes
from cryptography.hazmat.backends import default_backend
import os

def decrypt_file(encrypted_path, key, output_path):
    try:
        with open(encrypted_path, 'rb') as f:
            encrypted_data = f.read()

        iv = encrypted_data[:16]
        cipher_text = encrypted_data[16:]

        cipher = Cipher(algorithms.AES(key), modes.CBC(iv), backend=default_backend())
        decryptor = cipher.decryptor()

        # Decrypt the data
        data = decryptor.update(cipher_text) + decryptor.finalize()

        # Check for JPEG "magic bytes" (FF D8 FF) to verify decryption was successful
        if data.startswith(b'\xff\xd8\xff'):
            with open(output_path, 'wb') as f:
                f.write(data)
            print("✅ File Decrypted Successfully!")
        else:
            print("❌ Wrong Key! The decrypted data is corrupted (Not a valid JPEG).")

    except ValueError as e:
        # This will catch errors like an invalid key length (e.g., if key isn't 16, 24, or 32 bytes)
        print(f"❌ Error: {e}")
    except Exception as e:
        print(f"❌ An unexpected error occurred: {e}")

# MAIN
key_input = input("Enter decryption key: ")
key = key_input.encode()

# Note: For AES, the length of the string `key` MUST be exactly 16, 24, or 32 bytes.
encrypted_file = r"C:\Learning_Qiskit\Files\Encrypt\Zoro_Onepiece_Image_Encrypted.bin"
decrypted_file = r"C:\Learning_Qiskit\Files\Decrypt\Zoro_Onepiece_Image_Decrypted.jpg"

decrypt_file(encrypted_file, key, decrypted_file)
```
